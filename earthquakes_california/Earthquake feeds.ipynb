```python
import requests
url = "https://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/all_month.csv"
all_info = requests.get(url)
all_info
```

```python
import pandas as pd
df = pd.read_csv(url)
df
```

```python
pd.set_option('display.max_columns', None)
df
```

```python
california = df[df['place'].str.contains('CA')]
california
```

```python
columns_to_remove = ['magType', 'nst', 'gap', 'dmin', 'rms', 'net', 
                     'updated', 'horizontalError', 'depthError', 
                     'magError', 'magNst', 'status', 'locationSource', 'magSource']
df = california.drop(columns=columns_to_remove)
df
```

```python
df.to_csv('california_earthquakes.csv', index=False)
```
